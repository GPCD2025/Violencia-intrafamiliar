## Kmeans Clustering

Para el proyecto the propone implementar Kmeans 

## Paquetes usados

```python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler 
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
df_encoded = pd.DataFrame(encoder.fit_transform(df_selected),
                          columns=encoder.get_feature_names_out(df_selected.columns))

# 3. Scale the encoded data so that all features contribute equally
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

# 4.Implementación de Keans
k = 3  # numero de clusters
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
labels_kmeans = kmeans.fit_predict(X_scaled)

# calculo de silhouette score para calificar la calidad 
silhouette = silhouette_score(X_scaled, labels_kmeans)
print(f"Silhouette Score for K-Means: {silhouette:.4f}")

# 5. Visualizacion usando PCA y reduciendo la dimension a 2 para graficar.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_kmeans, cmap='viridis', s=50)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2") 
plt.title("K-Means Clusters Visualized with PCA")
plt.colorbar(label="Cluster")
plt.show()
```

## K-Modes Clustering 

implementación de K-modes.

```python
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df_selected = df_selected.astype(str)
```

```python
df_encoded = df_selected.apply(lambda x: LabelEncoder().fit_transform(x))
```

```python

km = KModes(n_clusters=3, init="Huang", n_init=1, verbose=1, max_iter=20)
df_encoded = df_encoded.fillna("Unknown") 
clusters = km.fit_predict(df_encoded)

df_encoded["clusters"] = clusters 
```

```python
sns.countplot(x=df_encoded["clusters"])
plt.title("Cluster Distribution")
plt.show()
```

```python
df_encoded.groupby("clusters").mean()
sns.boxplot(x="clusters", y="Estado Civil", data=df_encoded)
```

```python
df_encoded.groupby("clusters")["Estado Civil"].agg(lambda x: x.mode()[0])
```
